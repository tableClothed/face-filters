```python
import tensorflow as tf
import tensorflow.keras.layers as ly
import numpy as np
import pandas as pd

x = pd.read_csv('data/training.csv')
```

```python
x["Image"] = x["Image"].apply(lambda im: np.fromstring(im, sep=" "))

x["Image"]
```

```python
x = x.dropna()
```

```python
images = np.vstack(x["Image"].values) / 255.0
images = images.astype(np.float32)
images = np.array(images).reshape(-1, 96, 96, 1)
images[0][0][0]
```

```python
from sklearn.utils import shuffle

X = images
y = x[x.columns[:-1]].values
y = (y - 48) / 48 # scaling to [-1, 1]
X, y = shuffle(X, y, random_state=42)
y = y.astype(np.float32)
X
```

```python
model = tf.keras.models.Sequential()

model.add(ly.Conv2D(32, (5, 5), input_shape=(96, 96, 1), activation="relu"))
model.add(ly.MaxPooling2D((2,2)))

model.add(ly.Conv2D(63, (3, 3), activation="relu"))
model.add(ly.MaxPooling2D(2, 2))
model.add(ly.Dropout(0.1))

model.add(ly.Conv2D(128, (3, 3), activation="relu"))
model.add(ly.MaxPooling2D((2,2)))
model.add(ly.Dropout(0.3))

model.add(ly.Flatten())

model.add(ly.Dense(64, activation="relu"))
model.add(ly.Dense(128, activation="relu"))
model.add(ly.Dense(256, activation="relu"))
model.add(ly.Dense(64, activation="relu"))
model.add(ly.Dense(30))
```

```python
model.compile(optimizer="adam",
                loss="mean_squared_error",
                metrics=['accuracy'])
```

```python
model.fit(X, y, epochs=100, batch_size=200, verbose=1)
```
